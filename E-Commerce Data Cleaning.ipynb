```python
import pandas as pd
```

```python
df = pd.read_csv("OnlineRetail.csv", encoding='latin1')
```

```python
# First rows
df.head(20)
```

```python
# Shape
df.shape

# Info
df.info()

# Data types
df.dtypes
```

The dataset contains 541,909 records and 8 columns.
The numerical columns are Quantity, UnitPrice, and CustomerID, while the remaining columns are categorical.
The InvoiceDate column is incorrectly stored as an object and requires conversion to datetime format.
Missing values were identified in the Description and CustomerID columns, with CustomerID having a significant number of missing entries.

```python
df.isnull().sum()
```

```python
# Drop rows with missing Description
df = df.dropna(subset=['Description'])

# Check result
df.isnull().sum()
```

The Description column contained 1,454 missing values, which were removed because product descriptions are essential for analysis.
The CustomerID column had a large number of missing values (135,080), which were retained since they likely represent guest transactions and removing them would significantly reduce the dataset.

After removing rows with missing product descriptions, the dataset no longer contains null values in the Description column. The number of missing CustomerID values slightly decreased as some of those rows were also removed. CustomerID missing values were retained to preserve transaction data.

```python
#Check for duplicates
df.duplicated().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Check again
df.duplicated().sum()
```

The dataset contained 5,268 duplicate records, which were removed to ensure data accuracy and prevent inflated analysis results. Duplicate entries can distort metrics such as total sales and product demand.

```python
#Convert InvoiceDate to Datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Check
df.info()
```

```python
#Check negative quantity
df[df['Quantity'] < 0].shape
```

```python
# Keep only positive quantities
df = df[df['Quantity'] > 0]

# Check again
df[df['Quantity'] < 0].shape
```

```python
#Unit price check
df[df['UnitPrice'] <= 0].shape
```

```python
# Keep only valid prices
df = df[df['UnitPrice'] > 0]

# Check again
df[df['UnitPrice'] <= 0].shape
```

The dataset contained 584 records with zero or negative unit prices. These were removed as they represent invalid or non-revenue-generating transactions and would negatively impact sales analysis.

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
